# Torch implementation

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class BasicBlock(nn.Module):
    def __init__(self, in_channel, factor_channel = 1, stride = 1):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channel,in_channel * factor_channel, kernel_size = 3, padding = 1,stride = stride, bias = False)
        self.bnorm1 = nn.BatchNorm2d(in_channel * factor_channel, affine = False, eps = 1e-8, momentum = 0)
        self.conv2 = nn.Conv2d(in_channel * factor_channel,in_channel * factor_channel, kernel_size = 3, padding = 1, bias = False)
        self.bnorm2 = nn.BatchNorm2d(in_channel * factor_channel, affine = False,eps = 1e-8, momentum = 0)
        if factor_channel != 1 or stride != 1:
            self.downsample = nn.Conv2d(in_channel, in_channel * factor_channel,kernel_size = 1, stride = stride, bias = False)
            self.bnorm_skip = nn.BatchNorm2d(in_channel * factor_channel, affine = False, eps = 1e-8,momentum = 0)
        else:
            self.downsample = None
            
    def forward(self,x):
        conv1 = self.bnorm1(self.conv1(x))
        relu1 = F.relu(conv1)
        conv2_temp = self.conv2(relu1)
        conv2 = self.bnorm2(conv2_temp)
        if self.downsample is not None:
            out = self.bnorm_skip(self.downsample(x))
            out = out + conv2
        else:
            out = conv2 + x
        out = F.relu(out)
        return out
        
```

```python
NUM_CLASSES = 10
class ResNet18(nn.Module):
    def __init__(self, first_channel, nb_blocks = 2, nb_layers = 2):
        super().__init__()
        self.conv1 = nn.Conv2d(1, first_channel, kernel_size = 7, padding = 3, stride = 2,bias = False)
        self.bnorm1 = nn.BatchNorm2d(first_channel,eps = 1e-8, momentum = 0, affine = False)
        
        self.maxpool1 = nn.MaxPool2d(kernel_size = 2)
        list_blocks = []
        for l in range(nb_layers):
            list_blocks.append(BasicBlock(first_channel, 1,1))
        
        for i in range(1,nb_blocks):
            list_blocks.append(BasicBlock(2**(i-1)*first_channel, 2, stride = 2))
            for l in range(1,nb_layers):
                list_blocks.append(BasicBlock(2**i*first_channel,1,1))

        self.blocks = nn.Sequential(*list_blocks)
        # Here, we multiply by 4 since the remaining shape out of the last block is 2x2
        self.ln = nn.Linear(2**(nb_blocks-1)*first_channel*16, NUM_CLASSES, bias = False)

    def forward(self, x):
        conv1 = self.conv1(x)
        conv1_b = self.bnorm1(conv1)
        conv1_relu = F.relu(conv1_b)

        maxpool = self.maxpool1(conv1_relu)
        out_blocks = self.blocks(maxpool)
        B, C,_ ,_=out_blocks.shape
        out_blocks=out_blocks.view(B,-1)
        out = self.ln(out_blocks)

        return out
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
from tqdm import trange
batch_size = 64

model = ResNet18(64)
model = model.to(device)
loss_fn = nn.CrossEntropyLoss()
optim = torch.optim.SGD(model.parameters(), lr = 0.01)

losses = []
accuracies = []
trange = trange(1000)
for t in trange:
    indx = torch.randint(X_train.shape[0], (batch_size,))
    X = torch.tensor(X_train[indx].reshape(batch_size,1,28,28), dtype = torch.float)
    Y = torch.tensor(Y_train[indx])
    X, Y = X.to(device), Y.to(device)

    optim.zero_grad()
    out = model(X)
    loss = loss_fn(out, Y)
    acc = (torch.argmax(out, dim = -1) == Y).cpu().numpy().mean()
    losses.append(loss.item())
    accuracies.append(acc)

    loss.backward()
    optim.step()

    
    trange.set_description('Loss: %.3f, Acc: %.3f' %(loss.item(), acc))
    
```

# Numpy forward pass implementation

```python
import numpy as np

def forward_resnet(x):
    nb_batch, c_in,h_in, w_in = x.shape
    f_out, f_in, f_h, f_w = c1_w.shape

    l_conv1 = np.zeros((nb_batch, f_out, h_in//2, w_in//2))
    x_padded = np.pad(x, ((0,0),(0,0),(3,3),(3,3)))
    for i in range(l_conv1.shape[2]):
        for j in range(l_conv1.shape[3]):
            l_conv1[:,:,i,j] = np.sum( x_padded[:,np.newaxis,:, 2*i:2*i+7, 2*j:2*j+7]*c1_w[np.newaxis,:,:,:,:] , axis = (2,3,4))
    mean1 = np.mean(l_conv1 , axis = (0,2,3),keepdims = True)
    std1 = np.var(l_conv1, axis = (0,2,3),keepdims = True)
    l_bnorm1 = (l_conv1 -mean1)/(std1+1e-8)**0.5
    l_relu = np.maximum(l_bnorm1,0)

    l_maxpool = l_relu.reshape(nb_batch, f_out, l_conv1.shape[2]//2,2, l_conv1.shape[3]//2,2).max(axis =(3,5))
    
    mask_maxpool1 = (l_relu == l_maxpool.repeat(2,axis =2).repeat(2, axis =3)).astype('int')

    # Block 0: no stride for this block
    f_out, f_in, f_h, f_w = c01_w.shape
    maxpool_padded = np.pad(l_maxpool, ((0,0),(0,0),(1,1),(1,1)))
    l_conv01 = np.zeros((nb_batch, f_out,l_maxpool.shape[2], l_maxpool.shape[3]))
    for i in range(l_maxpool.shape[2]):
        for j in range(l_maxpool.shape[3]):
            l_conv01[:,:,i,j] = np.sum( maxpool_padded[:,np.newaxis, :, i:i+3,j:j+3] * c01_w[np.newaxis, :, :, :,:], axis = (2,3,4))
    
    mean01 = np.mean(l_conv01, axis = (0,2,3),keepdims = True)
    std01 = np.var(l_conv01, axis = (0,2,3),keepdims = True)
    l_bnorm01 = (l_conv01-mean01)/(std01+1e-8)**0.5
    l_relu01 = np.maximum(l_bnorm01 ,0)
    
    f_out, f_in, f_h, f_w = c02_w.shape
    l_conv02 = np.zeros((nb_batch, f_out, l_relu01.shape[2],l_relu01.shape[3]))
    l_relu01_padded = np.pad(l_relu01,((0,0),(0,0),(1,1),(1,1)))
    for i in range( l_relu01.shape[2]):
        for j in range(l_relu01.shape[3]):
            l_conv02[:,:,i,j] = np.sum(l_relu01_padded[:,np.newaxis, :, i:i+3, j:j+3] * c02_w[np.newaxis, :,:,:,:], axis = (2,3,4))
    
    mean02 = np.mean(l_conv02, axis = (0,2,3),keepdims = True)
    std02 = np.var(l_conv02, axis = (0,2,3),keepdims = True)
    l_bnorm02 = (l_conv02-mean02)/(std02+1e-8)**0.5
    
    l_res0 = l_maxpool + l_bnorm02
    l_block0 = np.maximum(l_res0, 0)
    
    # Block 1: no stride for this block
    f_out, f_in, f_h, f_w = c11_w.shape
    block0_padded = np.pad(l_block0, ((0,0),(0,0),(1,1),(1,1)))
    l_conv11 = np.zeros((nb_batch, f_out,l_block0.shape[2], l_block0.shape[3]))
    for i in range(l_block0.shape[2]):
        for j in range(l_block0.shape[3]):
            l_conv11[:,:,i,j] = np.sum( block0_padded[:,np.newaxis, :, i:i+3,j:j+3] * c11_w[np.newaxis, :, :, :,:], axis = (2,3,4))
    
    mean11 = np.mean(l_conv11, axis = (0,2,3),keepdims = True)
    std11 = np.var(l_conv11, axis = (0,2,3),keepdims = True)
    l_bnorm11 = (l_conv11-mean11)/(std11+1e-8)**0.5
    l_relu11 = np.maximum(l_bnorm11 ,0)
    
    f_out, f_in, f_h, f_w = c12_w.shape
    l_conv12 = np.zeros((nb_batch, f_out, l_relu11.shape[2],l_relu11.shape[3]))
    l_relu11_padded = np.pad(l_relu11,((0,0),(0,0),(1,1),(1,1)))
    for i in range(l_relu11.shape[2]):
        for j in range(l_relu11.shape[3]):
            l_conv12[:,:,i,j] = np.sum(l_relu11_padded[:,np.newaxis, :, i:i+3, j:j+3] * c12_w[np.newaxis, :,:,:,:], axis = (2,3,4))
    
    mean12 = np.mean(l_conv12, axis = (0,2,3),keepdims = True)
    std12 = np.var(l_conv12, axis = (0,2,3),keepdims = True)
    l_bnorm12 = (l_conv12-mean12)/(std12+1e-8)**0.5

    l_res1 = l_block0 + l_bnorm12

    l_block1 = np.maximum(l_res1, 0)

    # Block 2: stride=2 for this block, only first conv
    f_out, f_in, f_h, f_w = c21_w.shape
    block1_padded = np.pad(l_block1, ((0,0),(0,0),(1,1),(1,1)))
    l_conv21 = np.zeros((nb_batch, f_out,int(np.ceil(l_block1.shape[2]/2)), int(np.ceil(l_block1.shape[3]/2))))
    for i in range(l_conv21.shape[2]):
        for j in range(l_conv21.shape[3]):
            l_conv21[:,:,i,j] = np.sum( block1_padded[:,np.newaxis, :, 2*i:2*i+3,2*j:2*j+3] * c21_w[np.newaxis, :, :, :,:], axis = (2,3,4))
    
    mean21 = np.mean(l_conv21, axis = (0,2,3),keepdims = True)
    std21 = np.var(l_conv21, axis = (0,2,3),keepdims = True)
    l_bnorm21 = (l_conv21-mean21)/(std21+1e-8)**0.5
    l_relu21 = np.maximum(l_bnorm21 ,0)

    f_out, f_in, f_h, f_w = c22_w.shape
    l_conv22 = np.zeros((nb_batch, f_out, l_relu21.shape[2],l_relu21.shape[3]))
    l_relu21_padded = np.pad(l_relu21,((0,0),(0,0),(1,1),(1,1)))
    for i in range(l_relu21.shape[2]):
        for j in range(l_relu21.shape[3]):
            l_conv22[:,:,i,j] = np.sum(l_relu21_padded[:,np.newaxis, :, i:i+3, j:j+3] * c22_w[np.newaxis, :,:,:,:], axis = (2,3,4))
    
    mean22 = np.mean(l_conv22, axis = (0,2,3),keepdims = True)
    std22 = np.var(l_conv22, axis = (0,2,3),keepdims = True)
    l_bnorm22 = (l_conv22-mean22)/(std22+1e-8)**0.5

    l_resid2 = np.zeros((nb_batch, f_out, l_bnorm22.shape[2],l_bnorm22.shape[3]))
    for i in range(l_bnorm22.shape[2]):
        for j in range(l_bnorm22.shape[3]):
            l_resid2[:,:,i,j] = np.sum(l_block1[:,np.newaxis, :, 2*i, 2*j] *c2downsample_w[np.newaxis, :,:,0,0] ,axis = 2)
    
    resid_mean = np.mean(l_resid2, axis = (0,2,3),keepdims = True)
    resid_std = np.var(l_resid2, axis = (0,2,3),keepdims = True)
    l_resid2_bnorm = (l_resid2-resid_mean)/(resid_std+1e-8)**0.5
    
    l_res2 = l_resid2_bnorm + l_bnorm22
    l_block2 = np.maximum(l_res2, 0)
        
    # Block 3: no stride for this block
    f_out, f_in, f_h, f_w = c31_w.shape
    block2_padded = np.pad(l_block2, ((0,0),(0,0),(1,1),(1,1)))
    l_conv31 = np.zeros((nb_batch, f_out,l_block2.shape[2], l_block2.shape[3]))
    for i in range(l_block2.shape[2]):
        for j in range(l_block2.shape[3]):
            l_conv31[:,:,i,j] = np.sum( block2_padded[:,np.newaxis, :, i:i+3,j:j+3] * c31_w[np.newaxis, :, :, :,:], axis = (2,3,4))
    
    mean31 = np.mean(l_conv31 , axis = (0,2,3),keepdims = True)
    std31 = np.var(l_conv31, axis = (0,2,3),keepdims = True)
    l_bnorm31 = (l_conv31-mean31)/(std31+1e-8)**0.5
    l_relu31 = np.maximum(l_bnorm31 ,0)

    f_out, f_in, f_h, f_w = c32_w.shape
    l_conv32 = np.zeros((nb_batch, f_out, l_relu31.shape[2],l_relu31.shape[3]))
    l_relu31_padded = np.pad(l_relu31,((0,0),(0,0),(1,1),(1,1)))
    for i in range(l_relu31.shape[2]):
        for j in range(l_relu31.shape[3]):
            l_conv32[:,:,i,j] = np.sum(l_relu31_padded[:,np.newaxis, :, i:i+3, j:j+3] * c32_w[np.newaxis, :,:,:,:], axis = (2,3,4))
    
    mean32 = np.mean(l_conv32, axis = (0,2,3),keepdims = True)
    std32 = np.var(l_conv32, axis = (0,2,3),keepdims = True)
    l_bnorm32 = (l_conv32-mean32)/(std32+1e-8)**0.5

    l_res3 = l_block2 + l_bnorm32

    l_block3 = np.maximum(l_res3, 0)
    l_block3_reshaped = l_block3.reshape(nb_batch, -1)
    out = l_block3_reshaped.dot(l1.T)
    return out
```
